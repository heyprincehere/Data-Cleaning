### Step 5 Missing value treatments

```python
#choose the method of imputing missing value
#like mean,median, mode or KNNIputer
```

```python
for i in ["Polio","Income composition of resources"]:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
knn_imputer = KNNImputer(n_neighbors=5)
```

```python
for i in df.select_dtypes(include="number").columns:
    df[[i]] = knn_imputer.fit_transform(df[[i]])
```

### Step 6: Outliers treatments

```python
#decide wheather to do or not to do outlier treatment, if do then how?
    
```

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['GDP'])
```

```python
for i in ['GDP','Total expenditure',' thinness  1-19 years', ' thinness 5-9 years']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
for i in ['GDP','Total expenditure',' thinness  1-19 years', ' thinness 5-9 years']:
    sns.boxplot(df[i])
    plt.show()
```
